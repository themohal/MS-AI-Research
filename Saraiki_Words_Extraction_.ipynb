```python
from bs4 import BeautifulSoup as bs
import requests
import nltk
nltk.download('words')
from nltk.corpus import words
  
  
correct_words = words.words()
```

```python
spd = []
```

```python
for w in correct_words:
  url = "https://app.glosbe.com/en/skr/"+w
  response = requests.get(url)
  html = response.content
  soup = bs(html, "lxml")
  element = soup.find("h4", {"class":"dense","id":"translation-7040237068145291262"})
  if element!=None:
    spd.append([w,element.text.strip()])
    print(element.text.strip())
```
