```python
import nltk
nltk.download('words')
  
from nltk.corpus import words
word_list = words.words()
len(word_list)
```

```python
from bs4 import BeautifulSoup
import requests

page = requests.get('https://www.ijunoon.com/dictionary/tree-saraiki-meaning/')
soup = BeautifulSoup(page.content, 'html.parser')
print(soup.prettify()) ##shows source in html format
```

```python
import time
eng_word = []
first_word = []
second_word = []
count = 0
for i in word_list:
  eng_word.append(i)
  page = requests.get('https://www.ijunoon.com/dictionary/'+i+'-saraiki-meaning/')
  soup = BeautifulSoup(page.content, 'html.parser')
  d = soup.find_all("div", class_="m",text=True)
  if len(d)==2:
    count +=1
    print(count)
    for j in range(len(d)):
      
      if j==0:
        first_word.append(d[j].find('p').text)
      else:
        second_word.append(d[j].find('p').text)
      
  if len(d)>2:
    count +=1
    print(count)
    first_word.append(d[0].find('p').text)
    for j in range(1,len(d)):
      
      second_word.append([d[j].find('p').text])
      


  
```

```python
for i,j,k in zip(eng_word,first_word,second_word):
  print(i,j,k)
```
